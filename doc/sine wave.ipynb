```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# assume 44.1kHz
fs = 44100
amplitude = 1
```

```python
def get_sinewave(freq, frames, times):
    return amplitude * np.sin(2 * np.pi * freq * times)

def get_cosinewave(freq, frames, times):
    return amplitude * np.cos(2* np.pi * freq * times)

def get_time_array(start_frame, frames):
    return np.linspace(start_frame/fs, (start_frame + frames) / fs, frames)
```

```python
frames = 300

freq_C = 261.63

frame_chunk = int(fs/freq_C * 5) # a five peaks example

time_array = np.linspace(0, frame_chunk/fs, fs, endpoint=False)

sinewave = get_sinewave(freq_C, frame_chunk, time_array)

plt.figure(figsize=(20,10))
plt.plot(time_array, sinewave)
```

### Test amplitude attenuation

```python
time_array = np.linspace(0, frame_chunk/fs, fs, endpoint=False)

sinewave = get_sinewave(freq_C, frame_chunk, time_array) * np.linspace(1, 0, fs)

plt.figure(figsize=(20,10))
plt.plot(time_array, sinewave)
```

### Phase alignment between two sine waves with different frequencies

```python
freq_F = 349.234
frame_chunk = int(fs/freq_C * 5) # a five C peaks example

time_array = get_time_array(0, frame_chunk) 


wave_C = get_sinewave(freq_C, frame_chunk, time_array)
wave_F = get_sinewave(freq_F, frame_chunk, time_array)
```

#### Sine wave of C and F

```python
plt.figure(figsize=(20,10))
plt.plot(time_array, wave_C, color='y')
plt.plot(time_array, wave_F, color='b')
```

### Wave changed without alignment
A miss alignment between previous and current wave typically produces a unwanted 'popping' sound during frequency changed.

```python
#eg. transition from C to F
# eg. alignment start at next chunk of time array

prev_frame_chunk = int(fs/freq_C * 3) - 100 # eg. prev chunk produce a approx. 3 peaks C wave
prev_time_array = get_time_array(0, prev_frame_chunk)

prev_wave = get_sinewave(freq_C, prev_frame_chunk, prev_time_array)

next_frame_chunk = int(fs/freq_F * 3) # eg. next chunk produce an approx. 3 peaks F wave
next_time_array = get_time_array(prev_frame_chunk, next_frame_chunk)

next_wave = get_sinewave(freq_F, next_frame_chunk, next_time_array)

full_time_array = np.concatenate((prev_time_array, next_time_array), axis=0)


plt.figure(figsize=(20,10))
plt.plot(prev_time_array, prev_wave, color='y')
plt.plot(next_time_array, next_wave, color='b')
plt.plot([prev_time_array[-1], prev_time_array[-1]], [prev_wave[-1], next_wave[0]], color='r')

# plt.plot(full_time_array, np.concatenate((np.zeros(prev_frame_chunk), next_wave)), color='b')
```

### Smooth transition via cross fade

```python
wave_length_C = int(fs/freq_C * 2)
wave_length_F = int(fs/freq_F * 6)

time_array_C = get_time_array(0, wave_length_C) 
time_array_F = get_time_array(0, wave_length_F) 


wave_C = get_cosinewave(freq_C, wave_length_C, time_array_C)
wave_F = get_cosinewave(freq_F, wave_length_F, time_array_F)


plt.figure(figsize=(20,10))
plt.plot(time_array_C, wave_C, color='y')
plt.plot(time_array_F, wave_F, color='b')
```

```python
# # let the cross fade sound wave to be 0.01s
cross_frame = int(2 * fs)
# make sure last frame is a complete cycle to prevent cracking sound
cross_frame = math.floor(cross_frame - cross_frame % (fs/freq_F))

# max_wave_length = max(wave_length_C, wave_length_F)
fade = int(0.5 * cross_frame)
cross_env = np.concatenate((
#     np.linspace(0, 1, fade),
    np.ones(cross_frame - fade),
    np.linspace(1, 0, fade)
)
)
cross_time_array = get_time_array(0, cross_frame)
prev_wave = get_cosinewave(freq_C, cross_frame, cross_time_array)
next_wave = get_cosinewave(freq_F, cross_frame, cross_time_array)

plt.figure(figsize=(20,10))
plt.plot(cross_time_array, prev_wave, color='y')
plt.plot(cross_time_array, next_wave, color='b')

cross_wave = prev_wave * cross_env + next_wave * (1-cross_env)

plt.figure(figsize=(20,10))
plt.plot(cross_time_array, cross_wave)
```
